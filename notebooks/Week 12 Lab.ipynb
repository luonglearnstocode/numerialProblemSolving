## Exercise 1  
Use least-squares regression to fit a straight line to  
  
\begin{array}{c|c c}
x & 0 & 2 & 4 & 6 & 9 & 11 & 12 & 15 & 17 & 19 \\ 
\hline
y & 5 & 6 & 7 & 6 & 9 & 8 & 7 & 10 & 12 & 12 \\ 
\end{array}
  
Along with the slope and intercept, compute the standard error of the estimate and the correlation coefficient. Plot the data and the regression line. Then repeat the problem, but regress $x$ versus $y$ - that is, switch the variables. Interpret your results.

```python
# 1
x = array([0, 2, 4, 6, 9, 11, 12, 15, 17, 19])
y = array([5, 6, 7, 6, 9,  8, 7, 10, 12, 12])

figure()
plot(x, y, 'o')
xlim(-1, 21)
ylim(0, 13)
grid()
show()
```

```python
results = linregress(x, y)

figure()
xlim(-1, 21)
ylim(0, 13)
plot(x, y, 'o')
plot(x, results.intercept + results.slope*x, 'r')
grid()
show()
print(results)
```

```python
x = array([0, 2, 4, 6, 9, 11, 12, 15, 17, 19])
y = array([5, 6, 7, 6, 9,  8, 7, 10, 12, 12])


figure()
xlim(4, 13)
ylim(-1, 21)
plot(y, x, 'o')
grid()
show()
```

```python
results = linregress(y, x)

figure()
xlim(4, 13)
ylim(-1, 21)
plot(y, x, 'o')
plot(y, results.intercept + results.slope*y, 'r')
grid()
show()
print(results)
```

```python
x = array([ 2.5, 3.5, 5, 6, 7.5, 10, 12.5, 15, 17.5, 20])
y = array([ 13, 11, 8.5, 8.2, 7, 6.2, 5.2, 4.8, 4.6, 4.3])
results = linregress(log(x), log(y))

b = results.slope
a = exp(results.intercept)

figure()
# plot(log(x), log(y), 'o')
# plot(log(x), results.intercept + results.slope*log(x), 'r')
plot(x, y, 'o')
plot(x, a * x ** b)
grid()
show()

x = 9
logy = results.intercept + results.slope*log(x)
y = exp(logy)
print(y)
```

## Exercise 4


```python
c = array([ 0.5, 0.8, 1.5, 2.5, 4])
k = array([1.1, 2.4, 5.3, 7.6, 8.9]) 

c2 = c ** 2

figure()
plt.ylabel('$k$')
plt.xlabel('$c^2$')
plot(c2, k, 'o')
grid()
show()
```

```python
c = array([ 0.5, 0.8, 1.5, 2.5, 4])
k = array([1.1, 2.4, 5.3, 7.6, 8.9]) 

c2 = c ** 2

results = linregress(1/c2, 1/k)

figure()
plt.ylabel('$1/k$')
plt.xlabel('$1/c^2$')
plot(1/c2, 1/k, 'o')
plot(1/c2, results.intercept + results.slope * 1/c2, 'r')
grid()
show()

kmax = 1 / results.intercept
cs = results.slope / results.intercept
print(kmax)
print(cs)

c = 2
k = kmax * c**2 / (cs + c**2)
print(k)
```

## Exercise 5

```python
x = array([0.2, 0.5, 0.8, 1.2, 1.7, 2.0, 2.3])
y = array([500, 700, 1000, 1200, 2200, 2650, 3750])

figure()
plot(x, y, 'o')
grid()
show()
```

```python
def f(x, a, b, c):
    return a * x**2 + b * x + c


popt, pcov = curve_fit(f, x, y)

figure()
plot(x, y, 'o')
plot(x, f(x, *popt), 'g--')
grid()
show()
```
