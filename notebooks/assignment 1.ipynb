```python
def is_vowel(letter):
   if letter in "aeiouy":
     return(True)
   else:
     return(False)
```

# scope rules

```python
def increase(n):
    n += 1
    print(n)
    
n = 1 
increase(n) # print 2 - local n
print(n) # print 1 - function increase has no impact on n
```

# matplotlib

```python
x = np.array(range(1,11))
x
```

```python
y1 = x**2
y2 = x**3
plt.figure(1)
plt.plot(x, y1, "ro-", linewidth = 4, markersize = 12)
plt.plot(x, y2, "gd-", linewidth = 4, markersize = 12)
plt.xlabel("$X$")
plt.ylabel("$Y$")
```

# np.random

```python
# simulating rolling 10 dices, 1000000 times, plot the histogram of the sum of the dices for each turn
rolls = np.sum(np.random.randint(1,7,(10,1000000)), axis=0)
plt.figure(2)
plt.hist(rolls, normed = True);
```
