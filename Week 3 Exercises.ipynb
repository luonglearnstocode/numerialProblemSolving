# Problem set 2 - Approximations and round-off errors

### Exercise 8

```python
# the true value of f(n)
f_true = pi**4/90
n = 10000
```

#### calculate f(n) by computing the sum from k = 1 to 10,000 in single precision

```python
f_in_order = single(0)
for k in range(1, n+1):
    f_in_order += single(1 / k**4)
f_in_order
```

```python
# calculate true percent relative error
e_t_in_order = single((f_true - f_in_order) / f_true * 100)
e_t_in_order
```

#### calculate f(n) by computing the sum from k = 10,000 to 1 in single precision

```python
f_in_reverse_order = single(0)
for k in range(n, 0, -1):
    f_in_reverse_order += single(1 / k**4)
f_in_reverse_order
```

```python
# calculate true percent relative error
e_t_in_reverse_order = single((f_true - f_in_reverse_order) / f_true * 100)
e_t_in_reverse_order
```

```python
# compare the two true percent relative errors
e_t_in_reverse_order < e_t_in_order
```

#### From the results above, we could notice that calculating f(n) by computing the sum in the reverse order from 10,000 to 1 brings more precise result (lower true percent relative error).

# Problem set 3 - Graphical analysis

## Exercise 1c

#### First, we create the function y to avoid duplicating code when we need to calculate y for a specific value of x later.


```python
def function_y(x):
    return x**5 - x**2 + 2
```

#### Next, we plot the function with the domain [-2, 2].

```python
x = arange(-2, 2.01, 0.01)
y = function_y(x)
figure(1)
plot(x, y, 'g-')
xlabel("$x$")
ylabel("$y$")
xlim([-2.5, 2.5])
title("$y = x^5 - x^2 + 2$")
grid()
show()
```

#### From the graph, we can see the root at x = -1.0, a local maximum at x = 2.0, a local minimum at x = -2.0. However, there are also a local minimum and a local maximum between -1.0 and 1.0. In order to locate these points, we need to limit the values and add more points.

```python
figure()
x = arange(-1.000, 1.2, 0.01)
y = function_y(x)
plot(x,y)
grid()
show()
```

#### Now we can see clearer that a local maximum is close to 0.0, and a local minimum is around 0.7. Zooming in we obtain a local maximum at 0.0 and a local minimum at 0.74. After having all the point needed, we calculate the values of the function at these points to plot them in the graph.

```python
# The root of the function
x0 = -1.00
y0 = function_y(x0)
# The 2 local maximum
x_max1 = 0.0
y_max1 = function_y(x_max1)
x_max2 = 2.0
y_max2 = function_y(x_max2)
# The 2 local minimum
x_min1 = -2.0
y_min1 = function_y(x_min1)
x_min2 = 0.74
y_min2 = function_y(x_min2)
```

#### We plot these points back to figure 1 above

```python
figure(1)
plot(x0, y0, 'ro')
annotate('root', xy=(x0, y0), xytext=(-0.5, 10), arrowprops=dict(facecolor='black', shrink=0.05),)
plot(x_max1, y_max1, 'bo')
annotate('local max', xy=(x_max1, y_max1), xytext=(-0.2, -10), arrowprops=dict(facecolor='black', shrink=0.05),)
plot(x_max2, y_max2, 'bo')
annotate('local max', xy=(x_max2, y_max2), xytext=(2.3, 33), arrowprops=dict(facecolor='black', shrink=0.05),)
plot(x_min1, y_min1, 'yo')
annotate('local min', xy=(x_min1, y_min1), xytext=(-1.5, -33), arrowprops=dict(facecolor='black', shrink=0.05),)
plot(x_min2, y_min2, 'yo')
annotate('local min', xy=(x_min2, y_min2), xytext=(0.6, -10), arrowprops=dict(facecolor='black', shrink=0.05),)
show()
```

```python
x = arange(-5, 5, 0.1)
y1 = x**3
y2 = 4 - 2*x
figure()
plot(x, y1, 'r-', label='$y = x^3$')
plot(x, y2, 'g-', label='$y = 4 - 2x$')
legend()
grid()
show()
```

## Exercise 4a

#### We plot the function with a limit in the y axis to be able to see the graph with the asymptotes.

```python
x = arange(-4, 4.1, 0.001)
y = (2*x + 1) / (x - 3)
figure()
ylim([-10, 10])
plot(x, y)
grid()
show()
```

#### From this graph, we can notice the vertical asymptote at x = 3. Also, this function has a horizontal asymptote at y = 2. We annotate these lines in the graph above.

```python
axhline(y=2)
annotate('vertical asymptote', xy=(3, 5), xytext=(0, 8), arrowprops=dict(facecolor='black', shrink=0.05),)
annotate('horizontal asymptote', xy=(-3, 2), xytext=(-2, 5), arrowprops=dict(facecolor='black', shrink=0.05),);
```
