```python
from scipy.misc import factorial
x = 0.3*pi
y_true = format(cos(0.3*pi), '.8f')
sign = 1.0
power = 0.0
y = 0
for i in range(10):
    y += sign * x ** power / factorial(power)
    sign *= -1
    power += 2
    if (format(y, '.8f')) == y_true:
        print(i+1)
        break
print(y)
    
```
