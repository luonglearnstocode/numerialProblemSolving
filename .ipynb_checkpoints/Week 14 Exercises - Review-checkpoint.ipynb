### Problem 5
Study the function `newton()` found in module `scipy.optimize`. Use that function to find the roots of the problem 1. Use tolerance value of 0.01. Try different initial estimate values $x_0$. Compare the results to the previous results. What do you observe?

### Solution

```python
f1 = lambda x: x**2 - x - 1
f2 = lambda x: x**3 - 3 * sin(x)
f3 = lambda x: exp(x) - 2
f4 = lambda x: 1 - x**2 + sin(50 / (1 + x**2))

x = linspace(-2, 2, 1000)
fig, axes = subplots(1, 4, figsize=(12, 3), sharey=True)

for n, f in enumerate([f1, f2, f3, f4]):
    axes[n].plot(x, f(x), lw=1.5)
    axes[n].axhline(0, color='r')
    axes[n].grid()
    axes[n].set_ylim(-5, 5)
    axes[n].set_xticks([-2, -1, 0, 1, 2])
    axes[n].set_xlabel(r'$x$', fontsize=18)
axes[0].set_ylabel(r'$f(x)$', fontsize=18)
titles = [r'$f(x)=x^2-x-1$', r'$f(x)=x^3-3\sin(x)$', r'$f(x)=e^x-2$', r'$f(x)=\sin\left(50/(1+x^2)\right)+1-x^2$']
for n, title in enumerate(titles):
    axes[n].set_title(title)
```

## Exercise 10 - Root finding part II

### Problem 2
Determine the roots of $f(x) = −13 − 20x + 19x^2 − 3x^3$  
(a) Graphically  
(b) Using simple fixed-point iteration  
(c) Finding the roots of the polynomial  

### Solution

```python
f = lambda x: -13 - 20 * x + 19 * x**2 - 3 * x**3

# a - graphically
x = linspace(-3, 6, 1000)
plot(x, f(x), lw = 1.5)
plot(-0.4468, 0, 'ro')
plot(2.050, 0, 'ro')
plot(4.7303, 0, 'ro')
grid()
axhline(0, color='r')
show()
```

```python
fig, axes = subplots(1, 3, figsize=(15, 4), sharey=True)

axes[0].plot(x, f(x))
axes[0].axhline(0, color='r')
axes[0].grid()
axes[0].set_xlim(-0.45, -0.44)
axes[0].set_ylim(-0.5, 0.5)
axes[0].set_xlabel(r'$x$', fontsize=18)
axes[0].set_ylabel(r'$f(x)$', fontsize=18)
axes[0].plot(-0.4468, 0, 'ro')

axes[1].plot(x, f(x))
axes[1].axhline(0, color='r')
axes[1].grid()
axes[1].set_xlim(2.045, 2.055)
axes[1].set_ylim(-0.5, 0.5)
axes[1].set_xlabel(r'$x$', fontsize=18)
axes[1].plot(2.050, 0, 'ro')

axes[2].plot(x, f(x))
axes[2].plot(4.7303, 0, 'ro')
axes[2].axhline(0, color='r')
axes[2].grid()
axes[2].set_xlim(4.725, 4.735)
axes[2].set_ylim(-0.5, 0.5)
axes[2].set_xlabel(r'$x$', fontsize=18);
```

```python
# b - using simple fixed-point iteration
g1 = lambda x: (-13 + 19 * x**2 - 3 * x**3) / 20
g2 = lambda x: ((13 + 20 * x + 3 * x**3) / 19) ** (1/2)
g3 = lambda x: ((-13 - 20 * x + 19 * x**2) / 3) ** (1/3)
gs = [g1, g2, g3]

def fixedPoint(g, x0, tol, maxiter=1000):
    """
    simple fixed-point iteration algorithm
    """
    e = 1
    n = 0
    while (e > tol and n < maxiter):
        x = g(x0)
        e = abs(x - x0) # calculate error
        x0 = x
        n += 1
    return x0, n

print("Root found by fixedPoint:")

x0s = [-0.5, 2, 4.7]
i = 0

for x0 in x0s:
    x0, n = fixedPoint(gs[i], x0, tol = 1e-6)
    i += 1
    print("\tRoot {}:".format(i))
    print("\t\tRoot x0 = {0:.4f}".format(x0))
    print("\t\tNumber of steps n = {}".format(n))
```

```python
# c - using numpy.polynomial.polynomial.polyroots
from numpy.polynomial.polynomial import polyroots

r = polyroots([-13, -20, 19, -3])
print('Polyroot result: ', real(r))
```

### Problem 6
The equations  
  
\begin{align}
    sin(x) + 3\ cos(x) - 2 &= 0 \\
    cos(x) - 3\ sin(y) &= 0
\end{align}  
  
have a solution close the point (1, 1). Refine the solution numerically. Can you represent the solution and the equations graphically?

### Solution

```python
# Find the solution numerically
from scipy.optimize import fsolve

def f(x):
    f1 = sin(x[0]) + 3 * cos(x[0]) - 2
    f2 = cos(x[0]) - 3 * sin(x[1])
    return [f1, f2]

x0 = fsolve(f, [1 , 1])
print("Root:", x0)
```

## Exercise 11 - Optimization

### Problem 5
Employ the following methods to find the maximum of
$f(x) = 4x − 1.8x^2 + 1.2x^3 − 0.3x^4 \tag*{}$  
  
(a) Golden-section method $(x_l = −2.0, x_u = 4.0, \epsilon_s = 0.0001)$.  
(b) Parabolic interpolation $(x_0 = 1.74, x_1 = 2.0, x_2 = 2.5, iterations = 20)$.  
(c) Newton-Raphson method $(x_0 = 3, \epsilon_s  = 0.0001)$.  

### Solution

```python
f = lambda x : 4 * x - 1.8 * x**2 + 1.2 * x**3 - 0.3 * x**4
nf = lambda x : -f(x)

figure()
x = linspace(-4, 4, 1000)
plot(x, f(x))
axhline(0, color = 'r')
grid()
show()
```

```python
# a - Golden-section method
gr = (math.sqrt(5) + 1) / 2

def gss(f, a, b, tol=1e-10, maxiter = 1000):
    '''
    golden section search
    to find the minimum of f on [a,b]
    f: a strictly unimodal function on [a,b]
    '''
    n = 0
    c = b - (b - a) / gr
    d = a + (b - a) / gr 
    while abs(c - d) > tol and n < maxiter:
        if f(c) < f(d):
            b = d
        else:
            a = c
        
        # we recompute both c and d here to avoid loss of precision which may lead to incorrect results or infinite loop
        c = b - (b - a) / gr
        d = a + (b - a) / gr
        n += 1

    return ((b + a) / 2, n)

print("Maximum found by golden-section method:")
x0, n = gss(nf, -2, 4, tol = 0.0001)
print("\tRoot x0 = {0:.7f}".format(x0))
print("\tNumber of steps n = {}".format(n))
```

```python
# b - Parabolic interpolation

def parabolicInterpolation(f, x0, x1, x2, tol = 1e-8, maxiter = 100):
    n = 0
    while abs(x2 - x0) > tol and n < maxiter: 
        n += 1
        x3 =  (f(x0) * (x1**2 - x2**2) + f(x1) * (x2**2 - x0**2) + f(x2) * (x0**2 - x1**2)) \
                    / (2 * (f(x0) * (x1 - x2) + f(x1) * (x2 - x0) + f(x2) * (x0 - x1)))
        if x3 > x1:
            x0, x1 = x1, x3
        else:
            x1, x2 = x3, x1

    return x1, n

print("Maximum found by parabolic interpolation:")
x0, n = parabolicInterpolation(f, 1.74, 2, 2.5, maxiter = 20)
print("\tRoot x0 = {0:.7f}".format(x0))
print("\tNumber of steps n = {}".format(n))
```

```python
# c - Newton-Raphson method
def derivative(f, x, h = 1e-8):
    """ 
    Inputs:
       f = the function 
       x = point to estimate the derivative on
       h = small number, default value 1e-8
    Outputs:
       df = estimate of the derivative of f at point x
    """
    df = (f(x + h) - f(x)) / h
    return df

def secondDerivative(f, x, h = 1e-8):
    """ 
    Inputs:
       f = the function 
       x = point to estimate the 2nd derivative on
       h = small number, default value 1e-8
    Outputs:
       df = estimate of the 2nd derivative of f at point x
    """
    ddf = (f(x + 2 * h) - 2 * f(x + h) + f(x)) / h**2
    return ddf

def newtonSearch(f, x0, tol = 1e-10, maxiter = 10000):
    """ 
    Inputs:
       f = the function 
       x0 = initial estimate of a extremum
       tol = tolerance for the extremum value
       maxiter = maximum numbers of iterations
    Outputs:
       x0 = estimate for the extremum
       n = number of iterations
    """
    n = 0
    while(abs(derivative(f, x0)) > tol and n < maxiter):
        n += 1
        fpp = secondDerivative(f, x0)
        if fpp != 0:
            x0 = x0 - derivative(f, x0) / fpp
        else:
            x0 += tol
    return x0, n 

print("Maximum found by Newton-Raphson method:")
x0, n = newtonSearch(f, 3, tol = 1e-4)
print("\tRoot x0 = {0:.7f}".format(x0))
print("\tNumber of steps n = {}".format(n))
```

## Exercise 12 - Curve fitting

### Problem 2
Fit the following data with the power model $(y = ax^b)$. Use the resulting power equation to predict y at $x = 9.0$.  
  
\begin{array}{c|c c}
x & 2.5 & 3.5 & 5 & 6 & 7.5 & 10 & 12.5 & 15 & 17.5 & 20 \\ 
\hline
y & 13 & 11 & 8.5 & 8.2 & 7 & 6.2 & 5.2 & 4.8 & 4.6 & 4.3 \\ 
\end{array}

### Solution

```python
from scipy.stats import linregress

x = array([ 2.5, 3.5, 5, 6, 7.5, 10, 12.5, 15, 17.5, 20])
y = array([ 13, 11, 8.5, 8.2, 7, 6.2, 5.2, 4.8, 4.6, 4.3])
results = linregress(log(x), log(y))

b = results.slope
a = exp(results.intercept)

figure()
plot(x, y, 'o')
x1 = linspace(1, 25, 1000)
plot(x1, a * x1 ** b, label="$y$ = {0:.2f}$x$ ^ ({1:.2f})".format(a, b))
grid()

x0 = 9
y0 = a * x0**b
plot(x0, y0, 'ro')
legend()
show()

print('Model: y = {0:.2f}x ^ ({1:.2f})'.format(a, b))
print('At x = 9.0, y = ', round(y0,4 ))
```

### Problem 4
An investigator has reported the data tabulated below for an experiment to determine the growth rate of bacteria $k$ (per day), as a function of oxygen concentration $c \ (mg/L)$. It is known that such data can be modeled by the following equation:  
  
$$k = \frac{k_{max}c^2}{c_s + c^2}$$  
  
where $c_s$ and $k_{max}$ are parameters. Use a transformation to linearize this equation. Then use linear reqression to estimate $c_s$ and $k_{max}$ and predict the growth rate at $c = 2 (mg/L)$.  
  
\begin{array}{c|c c}
c & 0.5 & 0.8 & 1.5 & 2.5 & 4 \\ 
\hline
k & 1.1 & 2.4 & 5.3 & 7.6 & 8.9 \\ 
\end{array}
  

### Solution
Devide 1 by both sides, we get a linear relation between $\frac{1}{k}$ and $\frac{1}{c^2}$. 
  
$$\frac{1}{k} = \frac{c_s}{k_{max}} \frac{1}{c^2} + \frac{1}{k_{max}}$$  

```python
c = array([ 0.5, 0.8, 1.5, 2.5, 4])
k = array([1.1, 2.4, 5.3, 7.6, 8.9]) 

results = linregress(1/c**2, 1/k)
kmax = 1 / results.intercept
print('kmax =', round(kmax, 4))
cs = results.slope / results.intercept
print('cs =', round(cs, 4))

figure()
plt.ylabel('k$')
plt.xlabel('$c$')
plot(c, k, 'o')

c1 = linspace(0.2, 5, 1000)
plot(c1, c1**2 * kmax / (cs + c1**2), 'r')

c0 = 2
k0 = kmax * c0**2 / (cs + c0**2)
plot(c0, k0, 'ro')
print('At c = 2, k = ', round(k0,4 ))

grid()
show()
```

## Exercise 13 - Numerical Integration

### Problem 3
Evaluate the following integral:  
  
$$\int_{-2}^{4}(1 - x - 4x^3 + 2x^5)\ dx $$
  
(a) analytically  
(b) single application of the trapezoidal rule  
(c) composite trapezoidal rule, with $n = 2$ and $n = 4$  
(d) single application of Simpson’s 1/3 rule  
(e) Simpson’s 3/8 rule  
(f) scipy.optimize.quad function  
  
For each of the numerical estimates determine the percent relative true error based on the analytical solution.

### Solution

```python
f = lambda x : 1 - x - 4 * x**3 + 2 * x**5
a = -2
b = 4

# Graphical presentation
x = linspace(a, b, 1000)
y = f(x)

# Fill the integration area
fill_between(x, y, alpha = 0.5)
# Make another graph showing the function
x2 = linspace(-3, 5, 1000)
y2 = f(x2)
plot(x2, y2)
axhline(0, color = 'r')
grid()
```

```python
# a - analytically
import sympy as sp
from scipy.integrate import trapz, quad, simps

x = sp.symbols('x')
I_true = sp.integrate(f(x), (x, a, b)).evalf()
print('Using sympy.integrate')
print('\tI: {:.4f}'.format(I_true))
```

```python
# b - single application of the trapezoidal rule 
n = 1
x = linspace(a, b, n + 1)
y = f(x)

I = trapz(y, x)
error = abs(I - I_true) / I_true * 100

print('Using single application of the trapezoidal rule ')
print('\tI: {:.4f}'.format(I))
print('\tError: {:.4f} %'.format(error))
```

```python
# c - multiple-application trapezoidal rule, with n = 2 and n = 4
ns = [2, 4]

print('Using multiple-application trapezoidal rule ')
for n in ns:
    print('\tn = {}'.format(n))
    x = linspace(a, b, n + 1)
    y = f(x)
    I = trapz(y, x)
    error = abs(I - I_true) / I_true * 100
    print('\t\tI: {:.4f}'.format(I))
    print('\t\tError: {:.4f} %'.format(error))
```

```python
# d - single application of Simpson’s 1/3 rule
n = 2
x = linspace(a, b, n + 1)
y = f(x)

I = simps(y, x)
error = abs(I - I_true) / I_true * 100

print('Using single application of Simpson’s 1/3 rule ')
print('\tI: {:.4f}'.format(I))
print('\tError: {:.4f} %'.format(error))
```

```python
# e - Simpson’s 3/8 rule

def simpson38(f, a, b):
    x1, x2, x3, x4 = linspace(a, b, 4)
    h = x2 - x1
    return 3 / 8 * h * (f(x1) + 3 * f(x2) + 3 * f(x3) + f(x4))

I = simpson38(f, a, b)
error = abs(I - I_true) / I_true * 100

print('Using single application of Simpson’s 3/8 rule ')
print('\tI: {:.4f}'.format(I))
print('\tError: {:.4f} %'.format(error))
```

```python
# f - use scipy.integrate.quad 
I, error = quad(f, a, b)
print('Using scipy.integrate.quad  ')
print('\tI: {:.4f}'.format(I))
print('\tError: {:.4f} %'.format(error))
```

### Problem 5
Intergrate the following function both analytically and using Simpson’s rules, with $n = 4, 5, \dots$. When does the relative true error is less than 0.01%.

$$\int_{-3}^{5}(4x - 3)^3\ dx $$

```python
f = lambda x : (4 * x - 3)**3
a = -3
b = 5

# Graphical presentation
x = linspace(a, b, 1000)
y = f(x)

# Fill the integration area
fill_between(x, y, alpha = 0.5)
# Make another graph showing the function
x2 = linspace(-4, 6, 1000)
y2 = f(x2)
plot(x2, y2)
axhline(0, color = 'r')
grid()
```

```python
x = sp.symbols('x')
I_true = sp.integrate(f(x), (x, a, b)).evalf()
print('Using sympy.integrate')
print('\tI: {:.4f}'.format(I_true))
```

```python
n = 4
while True:
    print('\tn = {}'.format(n))
    x = linspace(a, b, n + 1)
    y = f(x)
    I = simps(y, x)
    error = abs(I - I_true) / I_true * 100
    print('\tI: {:.4f}'.format(I))
    print('\tError:', error)
    if error < 0.01:
        break
    n += 1
print('The error is less than 0.01% when n = ', n)
```
