### Exercise 1
The logarithmic function can be approximated with the following series:  
$log(x)  \approx (x-1) - \frac{1}{2}(x-1)^2 + \frac{1}{3}(x-1)^3 - \frac{1}{4}(x-1)^4 + \frac{1}{5}(x-1)^5 - ...\tag*{}$  
(a) Draw in the same graph the logarithmic function and its 5th order polynomial approximation. Use a suitable domain.  
(b) How much is the relative true error at $x = 2.0$ for the 5th order polynomial approximation?  
(c) Find out how many terms are needed to approximate the logarithmic function correctly at least 8 significant figures at x = 2.0.



### Solution
(a) We create an array of x with value from 0.01 to 3.0 at 0.01 a step.
We then plot the function log(x) and the 5th order polynomial approximation function in the same graph.

```python
x = arange(0.01, 3.0, 0.01)
y = log(x)
y_about_5 = (x - 1) - 1/2 * (x - 1)**2 + 1/3 * (x - 1)**3 - 1/4 * (x - 1)**4 + 1/5 * (x - 1)**5

figure('exercise 1a')
xlabel("$x$")
ylabel("$log(x)$")
plot(x, y, 'b', label='$y$')
plot(x, y_about_5, 'g', label='$y_{about}$')
legend()
grid()
show()
```

(b) At $x = 2.0$, we calculate the true value of log(x) and the 5th order polynomial approximation.  
Then we calculate the relative error and print out the result with the percentage format. 

```python
# b
x = 2.0
y_about = (x - 1) - 1/2 * (x - 1)**2 + 1/3 * (x - 1)**3 - 1/4 * (x - 1)**4 + 1/5 * (x - 1)**5
y = log(x)
relative_error = (y - y_about) / y
print("Relative error at x = 2.0: {0:.4f} %".format(relative_error * 100))
```

(c)  We start with x - 1 as the first term i = 1 of the approximation series. We then continue adding $\frac{(-1)^i}{i+1}(x - 1)^{i+1} $ to the approximation of y, and for each term added, we calculate the true error. We break the loop 
and return the number of terms needed when the absolute value of the true error is less than the stop criteria, which is based on the number of required significant figures.

```python
x = 2.0
y = log(x)
n = 4                 # Number of significant figures
es = 0.5*10**(-n)     # Stop criteria
y_about = x - 1.0     # First term of series approximation
i = 1                 # Number of terms
imax = 100000000      # Maximum iterations

while i < imax:
    # Add next term in series
    y_about = y_about + (-1)**(i) / (i + 1) * (x - 1)**(i + 1)
    
    i += 1
    
    # True error
    et = y - y_about
    if abs(et) <= es:
        break
        
print("Number of terms:     {0:1d}".format(i))
print("True error:          {0:.16f}".format(et))
```

### Exercise 3
The equilibrium equations of a truss joints can be written  
\begin{align*}
          -P_1 + P_2 - (1/ \sqrt{2}) P_3   &= 0 \\
                   (1/ \sqrt{2}) P_3 + P_4 &= 18 \\
                - P_2  - (1/ \sqrt{2}) P_5 &= 0 \\
                         (1/ \sqrt{2}) P_5 &=  12 \\
                   (1/ \sqrt{2}) P_5 + P_6 &= 0 \\
                 -P_4 - (1/ \sqrt{2})P_5   &= 0 
\end{align*}
where the units of $P_i$ are in $kN$  
(a) Write the equations in matrix form in Python  
(b) Solve these equations using any functions of your choice found in scipy.linalg package.  
(c) Rearrange the rows and columns so as to obtain a lower triangular coefficient matrix, and then solve them by forward substituion.


### Solution
(a) We can easily translate the system of linear equations into matrix form of matrix A of the coefficient from $P_1$ to $P_6 $and b because the equations are already simplified.  
(b) After getting matrix A and b, we use the imported solve() function to get the solution. We then check the answer by compare dot product of A and x with b.  
(c) First we arrange the matrix to the lower triangular form, though the diagonal is not yet 1.0 for all value.  
Then we normalize the matrixes by dividing each row with the value on the diagonal line.  
After getting the lower triangular matrix, we solve the equations by implementing forward substitution.
We get the solution with the same values that appear in the result of the original solution but in different order.

```python
# a
A = array([[-1, 1, -1/2**0.5, 0, 0, 0],
           [0, 0, 1/2**0.5, 1, 0, 0],
           [0, -1, 0, 0, -1/2**0.5, 0],
           [0, 0, 0, 0, 1/2**0.5, 0],
           [0, 0, 0, 0, 1/2**0.5, 1],
           [0, 0, 0, -1, -1/2**0.5, 0]])
b = array([0, 18, 0, 12, 0, 0])
# b
x = solve(A, b)
print("Solution: ",x)
print('Verify by A.x - b:')
print(around(dot(A, x) - b))
```

```python
#(c)
# matrix in lower triangular shape
A = array([[1/2**0.5, 0, 0, 0, 0, 0],
           [1/2**0.5, 1, 0, 0, 0, 0],
           [-1/2**0.5, 0, -1, 0, 0, 0],
           [-1/2**0.5, 0, 0, -1, 0, 0],
           [0, 0, 0, 1, 1/2**0.5, 0],
           [0, 0, 1, 0, -1/2**0.5, -1]])
b = array([12.0, 0, 0, 0, 18, 0])

n = len(b)
# normalize to get lower triangular matrix
for i in range(n):
    factor = A[i,i]
    A[i] = A[i] / factor
    b[i] = b[i] / factor
print("Lower triangular coefficient matrix")
print("A = ") 
print(A)
print("b = ", b)

# forwards substitution
b[0] = b[0] / A[0, 0]
for k in range(1, n):
    b[k] = b[k] - dot(A[k, 0:k], b[0:k])
print("Solution: ", b)
```
